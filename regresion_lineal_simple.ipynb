```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#Cargar archivo csv desde seaborn
df = sns.load_dataset(name='car_crashes')
df.head(5)
```

```python
#Imprimimos  el scatter plot entre la variable dependiente (total) e independiente (a)
#para observar el comportamiento en su dispersión
from turtle import color

sns.scatterplot(x='alcohol', y='total', color="blue", data=df)  
sns.scatterplot(x='speeding', y='total', color="red", data=df)
```

```python
#Declaramos las variables dependientes e independientes para la regresión lineal
#Vars_Indep = df[['alcohol', 'speeding']]
Vars_Indep=df[['alcohol']]
Var_Dep=df['total']
```

```python
#Se define moel como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Verificamos la función relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X = Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado  coef = a    intercept = b
model.__dict__
```

**Modelo mtemático y=2.0325063x + 5.857776154826299**

```python
#Predecimos los valores de total de accidentes a partir de la variable 'alcohol' x 
#y_pred= model.predict(X=df[['alcohol', 'speeding']])
y_pred= model.predict(X=df[['alcohol']])
y_pred
```

```python
#Insertamos la columna de predicciones en el DataFrame
df.insert(0, 'Predicciones', y_pred)
df
```

```python
#Visualizamos la gráfica comparativa entre el total real y el total predecido
#Regresa el menor error posible a partir de alcohol (funsión de accidentes en relación a alcohol)
sns.scatterplot(x='alcohol', y='total', color="blue", data=df)
sns.scatterplot(x='alcohol', y='Predicciones', color="red", data=df)
#sns.lineplot(x='alcohol', y='Predicciones', color="red", data=df)
```

```python
#Corroboramos cuál es el coeficiente de Determinación de nuestro modelo 
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```
